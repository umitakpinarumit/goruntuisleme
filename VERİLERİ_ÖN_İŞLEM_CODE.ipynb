```python
import os
import random
import shutil

# İşlem yapılacak klasörün yolu
klasor_yolu = r'C:\Users\ahmet\Desktop\bitirme\Images'

# Klasördeki .png dosyalarını listele
dosya_listesi = [dosya for dosya in os.listdir(klasor_yolu) if dosya.endswith('.png')]

# %20 test, %80 eğitim olarak bölme
test_orani = 0.2
dosya_sayisi = len(dosya_listesi)
test_sayisi = int(dosya_sayisi * test_orani)
egitim_sayisi = dosya_sayisi - test_sayisi

# Dosyaları karıştır
random.shuffle(dosya_listesi)

# Eğitim ve test setleri için klasörlerin yollarını belirle
egitim_klasor_yolu = os.path.join(klasor_yolu, 'train')
test_klasor_yolu = os.path.join(klasor_yolu, 'test')

# Klasörleri oluştur
os.makedirs(egitim_klasor_yolu, exist_ok=True)
os.makedirs(test_klasor_yolu, exist_ok=True)

# Eğitim setine dosyaları taşı
for dosya in dosya_listesi[:egitim_sayisi]:
    dosya_yolu = os.path.join(klasor_yolu, dosya)
    hedef_dosya_yolu = os.path.join(egitim_klasor_yolu, dosya)
    shutil.move(dosya_yolu, hedef_dosya_yolu)

# Test setine dosyaları taşı
for dosya in dosya_listesi[egitim_sayisi:]:
    dosya_yolu = os.path.join(klasor_yolu, dosya)
    hedef_dosya_yolu = os.path.join(test_klasor_yolu, dosya)
    shutil.move(dosya_yolu, hedef_dosya_yolu)

```

```python
from PIL import Image
import os

# İşlem yapılacak klasörün yolu
klasor_yolu = r'C:\Users\ahmet\Desktop\bitirme\Images\train'

# Yeni boyut
yeni_genislik = 224
yeni_yukseklik = 224

# Klasördeki tüm dosyaları listele
dosya_listesi = os.listdir(klasor_yolu)

for dosya in dosya_listesi:
    # Dosya yolunu oluştur
    dosya_yolu = os.path.join(klasor_yolu, dosya)
    
    # Dosyanın bir resim dosyası olduğunu kontrol et
    if dosya.endswith(('png')):
        # Resmi aç
        with Image.open(dosya_yolu) as img:
            # Resmi yeniden boyutlandır
            yeniden_boyutlandirilmis_img = img.resize((yeni_genislik, yeni_yukseklik))
            
            # Yeniden boyutlandırılmış resmi kaydet
            yeniden_boyutlandirilmis_dosya_yolu = os.path.join(klasor_yolu, f"{dosya}")
            yeniden_boyutlandirilmis_img.save(yeniden_boyutlandirilmis_dosya_yolu)
    else:
        print(f"{dosya} isimli dosya bir resim dosyası değil, atlanıyor.")
```

```python
import os
import shutil

# İşlem yapılacak klasörün yolu
klasor_yolu = r'C:\Users\ahmet\Desktop\bitirme\Images\train'

# Klasördeki .png dosyalarını listele
dosya_listesi = [dosya for dosya in os.listdir(klasor_yolu) if dosya.endswith('.png')]

for dosya in dosya_listesi:
    # Dosyanın ilk harfini al
    ilk_harf = dosya[0]
    
    # Hedef klasör yolunu belirle
    hedef_klasor_yolu = os.path.join(klasor_yolu, ilk_harf)
    
    # Hedef klasör yoksa oluştur
    if not os.path.exists(hedef_klasor_yolu):
        os.makedirs(hedef_klasor_yolu)
    
    # Dosyayı taşı
    dosya_yolu = os.path.join(klasor_yolu, dosya)
    hedef_dosya_yolu = os.path.join(hedef_klasor_yolu, dosya)
    shutil.move(dosya_yolu, hedef_dosya_yolu)
```
